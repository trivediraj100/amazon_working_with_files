```python
import csv

basic_qualifications=[]

with open('amazon.csv', 'r',encoding='utf-8') as csvFile:
    reader=csv.DictReader (csvFile)
    for row in reader:
        basic_qualifications.append(row['BASIC QUALIFICATIONS'])

def findKeyWordList (arr):
    arrlist=[]
    for i in arr:
        if 'Bachelor' in i or 'BS' in i or 'BA' in i:
            arrlist.append(i)

    return arrlist

bachelor_job_list=findKeyWordList (basic_qualifications)
print(len(bachelor_job_list))
```
